```python
import pandas as pd

df = pd.read_csv("01_data/processed/nhanes_analytic_clean.csv")

# Create Tableau-friendly export with readable labels
tableau_df = df.copy()

# Make risk tier readable
tableau_df["risk_tier"] = tableau_df["risk_tier"].astype(str)

# Add age groups for filtering
tableau_df["age_group"] = pd.cut(
    tableau_df["age"],
    bins=[19, 39, 59, 80],
    labels=["20-39", "40-59", "60-80"]
).astype(str)

# Make insurance readable
tableau_df["insurance_status"] = tableau_df["has_insurance"].map({
    1: "Insured", 0: "Uninsured"
})

# Select columns Tableau needs
tableau_export = tableau_df[[
    "id", "age", "age_group", "gender", "race_ethnicity",
    "education", "poverty_income_ratio", "income_group",
    "hba1c", "systolic_bp", "diastolic_bp",
    "total_cholesterol", "bmi", "waist_circumference",
    "told_diabetic", "told_hypertensive", "told_high_cholesterol",
    "undiagnosed_diabetes", "undiagnosed_prediabetes",
    "undiagnosed_hypertension", "undiagnosed_hyperlipidemia",
    "composite_risk_score", "risk_tier",
    "insurance_status", "usual_care_site",
    "er_visit_binary", "hosp_binary",
    "er_visits", "overnight_hosp"
]]

tableau_export.to_csv("01_data/processed/nhanes_tableau_export.csv", index=False)
print(f"✅ Tableau export saved: {tableau_export.shape[0]:,} rows × {tableau_export.shape[1]} columns")
print(f"Columns: {tableau_export.columns.tolist()}")
```
